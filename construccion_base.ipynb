```python
import pandas as pd
```

```python
circuits = pd.read_csv('project/f1_data/circuits.csv', na_values='\\N').drop(['location', 'name', 'location', 'country', 'lat', 'lng', 'url'], axis=1)
circuits
```

```python
drivers = pd.read_csv('project/f1_data/drivers.csv', na_values='\\N').drop(['forename', 'url', 'surname', 'number', 'code', 'dob', 'nationality'], axis=1)
drivers
```

```python
lap_times = pd.read_csv('project/f1_data/lap_times.csv', na_values='\\N').drop(['time'], axis=1)
lap_times
```

```python
pit_stops = pd.read_csv('project/f1_data/pit_stops.csv', na_values='\\N').drop(['time', 'milliseconds'], axis=1)
# renombrar lap a pitlap, duration a pitduration
pit_stops = pit_stops.rename(columns={'lap': 'pitlap', 'duration': 'pitduration'})
pit_stops
```

```python
qualifying = pd.read_csv('project/f1_data/qualifying.csv', na_values='\\N').drop(['constructorId', 'number'], axis=1)
qualifying = qualifying.rename(columns={'position': 'qposition'})
qualifying
```

```python
races = pd.read_csv('project/f1_data/races.csv', na_values='\\N')
races = races[['raceId', 'year', 'circuitId']]
races
```

```python
results = pd.read_csv('project/f1_data/results.csv', na_values='\\N').drop(['constructorId', 'number', 'positionText', 'positionOrder', 'rank', 'time', 'milliseconds', 'fastestLap', 'points'], axis=1)
results
```

```python
status = pd.read_csv('project/f1_data/status.csv', na_values='\\N')
status
```

Mergin tables

```python
# Fusion circuit with races
races = races.merge(circuits, on='circuitId')
races
```

```python
# Adding the number of laps each circuit has
#results.groupby(results['circuitId']).laps.max()
```

```python
races
```

```python
# fusion driver with results
results = drivers.merge(results, on='driverId', how='left')
```

```python
results
```

```python
pit_stops
```

```python
# fussion races with pit_stops
races = races.merge(pit_stops, on='raceId', how='left')
races
```

```python
# results with statusid
results = results.merge(status, on='statusId')
```

```python
results = results.drop(['statusId'], axis=1)
```

```python
# fusion races with results (one to many)
races = races.merge(results, on=['raceId', 'driverId'], how='left')
```

```python
races
```

```python
# changing names to grid -> startposition and position -> endposition

races = races.rename(columns={'grid': 'startposition', 'position': 'endposition'})
```

```python
# fusion races with lap_times (one to many)

races = races.merge(lap_times, on=['raceId','driverId'], how='left')
```

```python
# renaming in qualifying table position to qposition
qualifying = qualifying.rename(columns={'position': 'qposition'})
```

```python
qualifying
```

```python
# fussion races with qualifying
races = races.merge(qualifying, on=['raceId', 'driverId'], how='left')
```

```python
races = races[races['year'] >= 2019]
```

```python
races
```

```python
# Reordering columns
new_order = ['raceId','circuitId', 'driverId','resultId','qualifyId', 'year', 'circuitRef','driverRef', 'alt', 'stop',
       'pitlap', 'pitduration', 'startposition',
       'endposition', 'laps', 'fastestLapTime', 'fastestLapSpeed',
       'lap', 'position', 'milliseconds',  'qposition', 'q1', 'q2',
       'q3','status']
races = races[new_order]
```

```python
races
```

Export final data

```python
races.to_csv('RACES.csv', encoding='utf-8', index=False)
```
